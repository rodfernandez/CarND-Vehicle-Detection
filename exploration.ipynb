```python
import glob
import random

import cv2
import matplotlib.pyplot as plt
from skimage.feature import hog

% matplotlib inline
```

```python
# load training data
vehicles = glob.glob('training_data/vehicles/**/*.png', recursive=True)
non_vehicles = glob.glob('training_data/non-vehicles/**/*.png', recursive=True)
```

```python
# visualize vehicle images
vehicles_samples = random.sample(vehicles, k=16)
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(8, 8))

for vehicle, axis in zip(vehicles_samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(vehicle, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
    axis.imshow(image, cmap='gray')

fig.tight_layout()
plt.savefig('output_images/vehicles.jpg')
plt.show()
```

```python
# visualize non-vehicle images
non_vehicle_samples = random.sample(non_vehicles, k=16)
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(8, 8))

for vehicle, axis in zip(non_vehicle_samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(vehicle, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
    axis.imshow(image, cmap='gray')

fig.tight_layout()
plt.savefig('output_images/non_vehicles.jpg')
plt.show()
```

```python
# visualize spacial features
samples = vehicles_samples + non_vehicle_samples
# random.shuffle(samples)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
    resized = cv2.resize(image, (32, 32))
    axis.imshow(resized, cmap='gray')

fig.tight_layout()
plt.savefig('output_images/spacial_features.jpg')
plt.show()
```

```python
# visualize color features (RGB)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2RGB)
    x = [image[:,:,0].flatten(), image[:,:,1].flatten(), image[:,:,2].flatten()]
    axis.hist(x, bins=12, range=(0, 256), color=['r', 'g', 'b'])
    
fig.tight_layout()
plt.savefig('output_images/color_features_rgb.jpg')
plt.show()
```

```python
# visualize color features (HSV)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2HSV)
    x = [image[:,:,0].flatten(), image[:,:,1].flatten(), image[:,:,2].flatten()]
    axis.hist(x, bins=12, range=(0, 256), color=['r', 'g', 'b'])
    
fig.tight_layout()
plt.savefig('output_images/color_features_hsv.jpg')
plt.show()
```

```python
# visualize color features (YUV)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2YUV)
    x = [image[:,:,0].flatten(), image[:,:,1].flatten(), image[:,:,2].flatten()]
    axis.hist(x, bins=12, range=(0, 256), color=['r', 'g', 'b'])
    
fig.tight_layout()
plt.savefig('output_images/color_features_yuv.jpg')
plt.show()
```

```python
# visualize color features (LUV)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2LUV)
    x = [image[:,:,0].flatten(), image[:,:,1].flatten(), image[:,:,2].flatten()]
    axis.hist(x, bins=12, range=(0, 256), color=['r', 'g', 'b'])
    
fig.tight_layout()
plt.savefig('output_images/color_features_luv.jpg')
plt.show()
```

```python
# visualize HOG features (GRAY)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2GRAY)
    hog_features, hog_image = hog(image,
                                  orientations=18,
                                  pixels_per_cell=(8, 8),
                                  cells_per_block=(2, 2),
                                  block_norm='L2-Hys',
                                  visualise=True,
                                  transform_sqrt=False,
                                  feature_vector=False)
    axis.imshow(hog_image, cmap='gray')

fig.tight_layout()
plt.savefig('output_images/hog_features_gray.jpg')
plt.show()
```

```python
# visualize HOG features (L channel)
fig, axes = plt.subplots(nrows=8, ncols=4, figsize=(8, 16))

for sample, axis in zip(samples, axes.flatten()):
    image = cv2.cvtColor(cv2.imread(sample, cv2.IMREAD_COLOR), cv2.COLOR_BGR2LUV)
    hog_features, hog_image = hog(image[:,:,0],
                                  orientations=18,
                                  pixels_per_cell=(8, 8),
                                  cells_per_block=(2, 2),
                                  block_norm='L2-Hys',
                                  visualise=True,
                                  transform_sqrt=False,
                                  feature_vector=False)
    axis.imshow(hog_image, cmap='gray')

fig.tight_layout()
plt.savefig('output_images/hog_features_l_channel.jpg')
plt.show()
```
